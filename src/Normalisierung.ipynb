```python
import pandas as pd

def minmaxNormalize(df):

    if not df.isnull().values.any() :

        for i in range(0,df.columns.size):

            max = df.iloc[:,i].max()
            min = df.iloc[:,i].min()
            for j in range(0, df.iloc[:,i].size):

                    if not max-min == 0:
                        df.iloc[j,i] = (df.iloc[j,i] - min) / (max - min)
            print(df.iloc[:,i])
    else:
        print("contains null")

    return df


```

```python
unionMatrix_Class = pd.read_csv("../src/unionMatrix_Class.csv")
unionMatrix_Class = unionMatrix_Class.drop("sample_ID" , axis = 1)
labels = unionMatrix_Class["label"]
unionMatrix_Class = unionMatrix_Class.iloc[:,:-1]
minMaxNormedUnionMat = minmaxNormalize(unionMatrix_Class)
minMaxNormedUnionMat["label"]= labels
```

```python
import numpy as np
def stdNormalize(df):

    for col in df.columns:
        colMean = df[col].mean()
        colSTD = np.std(df[col])
        for i in range(len(df[col])):
            copy = df[col][i]
            df[col][i] = (copy - colMean)/colSTD

    return df
```

```python
stdNormedUnionMat = stdNormalize(unionMatrix_Class)
```

```python
stdNormedUnionMat["label"] = labels
```

```python
minMaxNormedUnionMat.to_csv(path_or_buf="./minMaxNormedUnionMat.csv")
```

```python
stdNormedUnionMat.to_csv(path_or_buf="./stdNormedUnionMat.csv")
```
