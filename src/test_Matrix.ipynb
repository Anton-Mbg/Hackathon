```python
import numpy as np
import pandas as pd

level = 'class'
i = 'CAD'
df1 = pd.read_csv('../HackathonMicrobiomeData/'+ i +'/' + level + i + '_train.csv')

df2 = pd.read_csv('../src/DataMatrices/Class/unionMatrix_Class.csv')

# Replace the 1 values (sick) with the correct value of the sickness (3 for CAD)
df1.loc[df1['label'] == 1, 'label'] = 3

# Test if union_matrix is correct (it is)
for sample in df1['sample_ID'].values:
    if(np.array_equal(df1[df1['sample_ID']==sample].values, df2[df2['sample_ID']==sample].values)):
        print('equal', sample)
    else:
        print('NOT equal', sample)
```
