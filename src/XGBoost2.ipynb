```python
from sklearn.metrics import accuracy_score
import xgboost as xgb
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import f1_score, matthews_corrcoef

#df = pd.read_csv('../HackathonMicrobiomeData/CAD/ClassCAD_train.csv')
#df = pd.read_csv('../src/SMOTEclass.csv')
df = pd.read_csv("DataMatrices/order/unionMatrix_TreeBasedClassifier_Normalized_order.csv")

Y = np.array(df[["label"]]).ravel()
#X = df.drop(["label"], axis = 1)
X = df.drop(["sample_ID", "label"], axis = 1)

# X_test = X.iloc[:250]
# print(X_test.shape)
# Y_test = Y[:250]
# print(Y_test.shape)
# X_train = X.iloc[251:]
# print(X_train.shape)
# Y_train = Y[251:]
# print(Y_train.shape)

print(df.columns)


# Convert the dataframe to a DMatrix
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.35, random_state=0)

dtrain = xgb.DMatrix(data=x_train, label=y_train)
dtest = xgb.DMatrix(data=x_test, label=y_test)

#dtrain = xgb.DMatrix(data=X_train, label=Y_train)
#dtest = xgb.DMatrix(data=X_test, label=Y_test)

# Create the parameter dictionary for the XGBoost model
#params = {'objective': 'reg:squarederror', 'eval_metric': 'rmse'}
params = {
    'objective': 'multi:softmax',
    'max_depth': 20,
    'learning_rate': 0.1,
    'n_estimators': 5,
    'seed': 123,
    'num_class': 5 # specify the number of classes
}

# Create the XGBoost model
model = xgb.train(params, dtrain)

# Make predictions on the test set
predictions = model.predict(dtest)

# Evaluate the model
#accuracy = accuracy_score(y_test, predictions)
accuracy = accuracy_score(y_test, predictions)
print('Accuracy:', accuracy)

print("MCC", matthews_corrcoef(y_test,predictions))
```
